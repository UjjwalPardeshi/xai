```python
import os
import json
import cv2
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt

class TuSimpleLaneDataset(Dataset):
    def __init__(self, label_json_path, base_dir, img_size=(360, 640), transform=None):
        """
        Args:
            label_json_path (str): Path to the label JSON file (e.g., label_data_0313.json)
            base_dir (str): Root path to the dataset folder containing 'clips/'
            img_size (tuple): Desired output image size (height, width)
            transform (callable, optional): Optional transform to be applied on sample.
        """
        self.base_dir = base_dir
        self.img_size = img_size
        self.transform = transform
        
        # Load label data
        self.samples = []
        with open(label_json_path, 'r') as f:
            for line in f:
                item = json.loads(line.strip())
                self.samples.append(item)

    def __len__(self):
        return len(self.samples)

    def lanes_to_mask(self, lanes, h_samples):
        """
        Create binary mask from lane annotation points.
        Returns: np.array: Binary mask of shape (height, width)
        """
        mask = np.zeros(self.img_size, dtype=np.uint8)

        for lane in lanes:
            points = [(x, y) for x, y in zip(lane, h_samples) if x != -2]
            if len(points) < 2:
                continue
            
            # Scale points according to resized image if needed
            orig_h, orig_w = 720, 1280  # TuSimple standard
            scale_x = self.img_size[1] / orig_w
            scale_y = self.img_size[0] / orig_h
            
            scaled_points = [(int(x*scale_x), int(y*scale_y)) for (x, y) in points]
            pts = np.array(scaled_points, np.int32).reshape((-1, 1, 2))
            cv2.polylines(mask, [pts], isClosed=False, color=255, thickness=10)
        return mask

    def __getitem__(self, idx):
        sample = self.samples[idx]
        raw_file = sample['raw_file']  # e.g. clips/0313-1/10000/20.jpg
        lanes = sample['lanes']
        h_samples = sample['h_samples']
        
        # Load image
        img_path = os.path.join(self.base_dir, raw_file)
        image = cv2.imread(img_path)
        if image is None:
            raise FileNotFoundError(f"Image not found: {img_path}")
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        # Resize image
        image = cv2.resize(image, (self.img_size[1], self.img_size[0]))
        
        # Create lane mask
        mask = self.lanes_to_mask(lanes, h_samples)
        
        # Apply transforms if any
        if self.transform:
            augmented = self.transform(image=image, mask=mask)
            image = augmented['image']
            mask = augmented['mask']
        
        # Convert to FloatTensor and normalize [0,1]
        image = torch.from_numpy(image).permute(2, 0, 1).float() / 255.0  # 3xHxW
        mask = torch.from_numpy(mask).unsqueeze(0).float() / 255.0        # 1xHxW
        
        return image, mask

# ------------------------
# Example usage:
# ------------------------

# Set dataset root folder and label path:
DATASET_ROOT = 'TuSimple/TUSimple/train_set'   # Adjust this to your dataset root
LABEL_JSON = os.path.join(DATASET_ROOT, 'label_data_0313.json')

dataset = TuSimpleLaneDataset(LABEL_JSON, DATASET_ROOT)
data_loader = DataLoader(dataset, batch_size=4, shuffle=True, num_workers=2)

# Example visualization function:
def visualize_batch(images, masks):
    batch_size = images.size(0)
    plt.figure(figsize=(16, 4 * batch_size))
    for i in range(batch_size):
        img = images[i].permute(1, 2, 0).numpy()
        mask = masks[i].squeeze(0).numpy()
        plt.subplot(batch_size, 2, 2*i+1)
        plt.imshow(img)
        plt.title('Image')
        plt.axis('off')
        plt.subplot(batch_size, 2, 2*i+2)
        plt.imshow(mask, cmap='gray')
        plt.title('Lane Mask')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

for imgs, masks in data_loader:
    visualize_batch(imgs, masks)
    break
```

```python
import torch
import segmentation_models_pytorch as smp

# Make sure your TuSimpleLaneDataset definition is already in the notebook/script.
# If you've defined the class as above, just continue.
```

```python
DATASET_ROOT = 'TuSimple/TUSimple/train_set'   # Adjust this path to your dataset root
LABEL_JSON = os.path.join(DATASET_ROOT, 'label_data_0313.json')

dataset = TuSimpleLaneDataset(LABEL_JSON, DATASET_ROOT)
data_loader = DataLoader(dataset, batch_size=4, shuffle=True, num_workers=0)  # num_workers=0 for Windows/CPU use
```

```python
# Create a U-Net with a pretrained ResNet34 encoder
model = smp.Unet(
    encoder_name='resnet34',        # Backbone options: 'resnet18', 'resnet50', etc.
    encoder_weights='imagenet',     # Use ImageNet-pretrained weights
    in_channels=3,                  # RGB input as in your DataLoader
    classes=1,                      # Binary mask output (single channel)
    activation=None                 # We use BCEWithLogitsLoss, so leave as None
)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
```

```python
import torch.optim as optim

optimizer = optim.Adam(model.parameters(), lr=1e-3)
criterion = torch.nn.BCEWithLogitsLoss()  # Use BCE with logits for binary segmentation
```

```python
num_epochs = 5  # Start with a few epochs to check setup
for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    for imgs, masks in data_loader:
        imgs = imgs.to(device)
        masks = masks.to(device)
        optimizer.zero_grad()
        logits = model(imgs)  # [B,1,H,W]
        loss = criterion(logits, masks)
        loss.backward()
        optimizer.step()
        total_loss += loss.item() * imgs.size(0)
    print(f'Epoch {epoch+1}, Loss: {total_loss / len(dataset):.4f}')
```

```python
import matplotlib.pyplot as plt

model.eval()
with torch.no_grad():
    imgs, masks = next(iter(data_loader))
    imgs = imgs.to(device)
    masks = masks.to(device)
    logits = model(imgs)
    preds = torch.sigmoid(logits) > 0.5  # Apply sigmoid and threshold

    for i in range(imgs.size(0)):
        img = imgs[i].cpu().permute(1, 2, 0).numpy().clip(0, 1)
        gt_mask = masks[i].cpu().squeeze().numpy()
        pred_mask = preds[i].cpu().squeeze().numpy()
        plt.figure(figsize=(10, 3))
        plt.subplot(1, 3, 1); plt.imshow(img); plt.title('Image'); plt.axis('off')
        plt.subplot(1, 3, 2); plt.imshow(gt_mask, cmap='gray', vmin=0, vmax=1); plt.title('Ground Truth'); plt.axis('off')
        plt.subplot(1, 3, 3); plt.imshow(pred_mask, cmap='gray', vmin=0, vmax=1); plt.title('Prediction'); plt.axis('off')
        plt.show()
```
